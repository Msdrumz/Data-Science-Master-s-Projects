```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder



store_data = pd.read_csv('/Users/peytonbailey/WGU Data Science Masters/D599- Data Preparation and Exploration/Task 3/Megastore_Dataset_Task_3 3.csv')
```

```python
print(store_data.columns)
print(store_data['Region'].unique())
print(store_data['PaymentMethod'].unique())
```

```python
print(store_data['CustomerOrderSatisfaction'].unique())
print(store_data['OrderPriority'].unique())
```

```python

# Ordinal Encoding for OrderPriority
order_priority_encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])
store_data['OrderPriority'] = order_priority_encoder.fit_transform(store_data[['OrderPriority']])

# Ordinal Encoding for CustomerOrderSatisfaction
satisfaction_encoder = OrdinalEncoder(categories=[['Prefer not to answer','Very Dissatisfied', 'Dissatisfied', 'Satisfied', 'Very Satisfied' ]])
store_data['CustomerOrderSatisfaction'] = satisfaction_encoder.fit_transform(store_data[['CustomerOrderSatisfaction']])

# Label Encoding for Payment Method
paymentmethod_encoder = LabelEncoder()
store_data['PaymentMethod'] = paymentmethod_encoder.fit_transform(store_data['PaymentMethod'])

#Label Encoding for Region
region_encoder = LabelEncoder()
store_data['Region'] = region_encoder.fit_transform(store_data['Region'])

```

```python
print(store_data['Region'].unique())
print(store_data['PaymentMethod'].unique())
print(store_data['CustomerOrderSatisfaction'].unique())
print(store_data['OrderPriority'].unique())


```
